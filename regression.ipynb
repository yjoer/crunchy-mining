```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("data/output.csv")
```

```python
df.info()
```

```python
df.drop(columns=['ChgOffDate'], inplace=True)
```

```python
#Encode Data
encoder = OrdinalEncoder()

categorical_cols = ["City", "State", "BankState", "ApprovalDate", "DisbursementDate", "Industry", "Active"]
df[categorical_cols] = encoder.fit_transform(df[categorical_cols])

# Print the updated dataset
df.head()
```

```python
def metrics_fusion(y_pred, y):
    y_pred_inv = np.expm1(y_pred)
    y_inv = np.expm1(y)

    a = mean_absolute_error(y_inv, y_pred_inv)
    b = np.mean(np.abs((y_inv - y_pred_inv) / y_inv)) * 100
    c = np.sqrt(mean_squared_error(y_inv, y_pred_inv))
    d = mean_squared_error(y_inv, y_pred_inv)
    e = r2_score(y_inv, y_pred_inv)

    return {"mae": a, "mape": b, "rmse": c, "mse": d, "r2": e}
```
